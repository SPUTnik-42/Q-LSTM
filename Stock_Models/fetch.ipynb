```python
import pandas as pd
import yfinance as yf
stocks = ['^GSPC','^FTSE', '^N100', '^STOXX50E', '^HSI']
for i in stocks:
    data = pd.DataFrame()
    data = yf.download(tickers=i, period='2y', interval='1h')
    data.drop(columns=['Volume', 'Adj Close'], inplace=True, errors='ignore')
    data.to_csv('./Data/Stock/'+i+'.csv')
```
